```python
import statsmodels.api as sm
import pandas as pd
import matplotlib.pyplot as plt
from plotnine import ggplot, geom_point, aes, labs, scale_x_continuous, scale_y_continuous, guides
import numpy as np
```

**Task 1**

In this specific plot, we can see that the variables being mapped are y = life expectancy to x = income. More so, they created a bubble plot tha has a color fill pending the continent they are in. The size of each bubble could allude to the size of the population that comes from that region, which may be the country. There's also a label in the background with 2010 indicating that this dataset pertains to life expectancy based on income for that year. Points are a little spread out, but seem to be somewhat linear.

```python
q1_data = pd.read_csv("/content/q1data.csv")
q1_data.head()
```

```python
(ggplot(q1_data, aes(x='income', y='life_exp', color='four_regions', size='population')) +
 geom_point() +
 labs(title='2010',
      x='Income',
      y='Life Expectancy'))
```

```python
(ggplot(q1_data, aes(x='income', y='life_exp', color='four_regions', size='population')) +
 geom_point() +
 scale_x_log10(
     limits=[0, 200000],
     breaks=[500, 1000, 2000, 4000, 8000, 16000, 32000, 64000]) +
 scale_y_continuous(
        limits=[0, 90],
        breaks=[20, 30, 40, 50, 60, 70, 80, 90]
        ) +
 labs(title='2010',
      x='Income',
      y='Life Expectancy'))
```

FINAL PLOT

```python
(ggplot(q1_data, aes(x='income', y='life_exp', color='four_regions', size='population')) +
 geom_point() +
 scale_x_continuous(
     trans='log2',
     limits=[400, 140000],
     breaks=[500, 1000, 2000, 4000, 8000, 16000, 32000, 64000],
     labels=['500', '1000', '2000', '4000', '8000', '16000', '32000', '64000']
     ) +
 scale_y_continuous(
     limits=[10,90],
     breaks=[20, 30, 40, 50, 60, 70, 80, 90]
     ) +
 labs(title='2010',
      x='Income',
      y='Life Expectancy')
      + guides(size = False))
```

**Task 2**


In this specific plot, we can see that the variables being mapped are y = imports by % of GDP to x = exports by % of GDP. Similarly, another bubble plot was created such that the color fill represents the continent they are in while the size accounts the energy usage of that country. There's also a label in the background with 1997 indicating that this dataset pertains to the Imports by % of GDP based on the Exports by % of GDP. Points are very close together and there seems to be a positive linear association.

```python
q2_data = pd.read_csv("/content/q2data.csv")
q2_data.head()
```

```python
q2_data['energy'] = pd.to_numeric(q2_data['energy'], errors='coerce')
```

```python
(ggplot(q2_data, aes(x='exports', y='imports', color='four_regions', size='energy')) +
 geom_point() +
 labs(title='1997',
      x='Exports (% of GDP)',
      y='Imports (% of GDP)'))
```

FINAL PLOT

```python
(ggplot(q2_data, aes(x='exports', y='imports', color='four_regions', size='energy')) +
 geom_point() +
 scale_x_continuous(
     limits=[0, 240],
     breaks=[20, 40, 60, 80, 100, 120, 140, 160, 180, 200, 220],
     labels=['20', '40', '60', '80', '100', '120', '140', '160', '180', '200', '220']
     ) +
 scale_y_continuous(
     limits=[0,500],
     breaks=[50, 100, 150, 200, 250, 300, 350, 400]
     ) +
 labs(title='1997',
      x='Exports (\\% of GDP)',
      y='Imports (\\% of GDP)')
      + guides(size = False))
```
